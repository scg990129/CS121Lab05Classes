## Lab: Lab: Composition

In this lab you’ll practice object-oriented design using composition (classes made from other classes). You will build a Point class, then a Square built only from Points, then a Cube built only from Squares.

## Test Case 1: Point Class

```python
from Point import Point
import math

# Create a Point object
point1 = Point(3,3)
origin = Point(0,0)
pointInYAxis = Point(0,5)
pointInXAxis = Point(1,0)

print("Test Case 1: Point Class")
print("Origin Point:", origin)
print("Point 1:", point1)
print("Point in Y-axis:", pointInYAxis)
print("Point in X-axis:", pointInXAxis)

# Calculate the distance from the origin (0, 0)
print("Distance from origin:", point1.distance(origin))

# Slope
print("slope from origin:", point1.slope(origin))
print(f"Slope from y-axis (Vertical): {origin.slope(pointInYAxis)}") # Should be inf
print(f"Slope from x-axis (Horizontal): {origin.slope(pointInXAxis)}") # Should be 0

print("Test Case 1 completed successfully.")
```

## Test Case 2: Square Class

```python
from Point import Point
from Square import Square
import math

# Create Square objects
square1 = Square(Point(0, 0), Point(2, 0), Point(2, 2), Point(0, 2))
square2 = Square(Point(1, 1), Point(3, 1), Point(3, 3), Point(1, 3))

print("\nTest Case 2: Square Class")
print("Square 1:", square1)
print("Square 2:", square2)

# Calculate area and perimeter
print("Square 1 Area:", square1.area())
print("Square 1 Perimeter:", square1.perimeter())
print("Square 2 Area:", square2.area())
print("Square 2 Perimeter:", square2.perimeter())

print("Test Case 2 completed successfully.")
```
